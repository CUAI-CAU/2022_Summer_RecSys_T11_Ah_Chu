```python
import pandas as pd
import ast
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('FINAL_META_DATA.csv')
```

```python
df['description_embedding'] = df['description_embedding'].map(lambda x : ast.literal_eval(x))
df['category_embedding'] = df['category_embedding'].map(lambda x : ast.literal_eval(x))
```

```python
df.head(3)
```

```python
des_array = np.zeros((6186,768))
for i in range(len(df)):
    des_array[i,:] = np.array(df.loc[i,'description_embedding'])
    
cat_array = np.zeros((6186,98))
for i in range(len(df)):
    cat_array[i,:] = np.array(df.loc[i,'category_embedding'])
```

```python
des_cosine_sim = cosine_similarity(des_array, des_array)
```

```python
des_cosine_sim
```

```python
cat_cosine_sim = cosine_similarity(cat_array, cat_array)
```

```python
cat_cosine_sim
```

```python
asinlist = df['asin'].tolist()
```

```python
des_cosine_sim_df = pd.DataFrame(des_cosine_sim, index = asinlist, columns= asinlist)
```

```python
des_cosine_sim_df
```

```python
cat_cosine_sim_df = pd.DataFrame(cat_cosine_sim, index = asinlist, columns= asinlist)
```

```python
cat_cosine_sim_df
```

```python
des_cosine_sim_df.to_csv('description_cos_sim_by_asin.csv')
```

```python
cat_cosine_sim_df.to_csv('category_cos_sim_by_asin.csv')
```
