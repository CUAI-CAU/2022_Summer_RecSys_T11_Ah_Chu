```python
train_set = cornac.data.Dataset.from_uir(train.itertuples(index=False), seed=SEED)

print('Number of users: {}'.format(train_set.num_users))
print('Number of items: {}'.format(train_set.num_items))
```

```python
sset = cornac.data.Dataset.from_uir(data. itertuples(index=False), seed=SEED)

print('Number of users: {}'.format(sset.num_users))
print('Number of items: {}'.format(sset.num_items))
```

```python
test_set = cornac.data.Dataset.from_uir(test.itertuples(index=False), seed=SEED)

print('Number of users: {}'.format(test_set.num_users))
print('Number of items: {}'.format(test_set.num_items))
```

```python
valid_set = cornac.data.Dataset.from_uir(valid.itertuples(index=False), seed=SEED)

print('Number of users: {}'.format(valid_set.num_users))
print('Number of items: {}'.format(valid_set.num_items))
```

```python
bivae = cornac.models.BiVAECF(
    k=LATENT_DIM,
    encoder_structure=ENCODER_DIMS,
    act_fn=ACT_FUNC,
    likelihood=LIKELIHOOD,
    n_epochs=NUM_EPOCHS,
    batch_size=BATCH_SIZE,
    learning_rate=LEARNING_RATE,
    seed=SEED,
    use_gpu=torch.cuda.is_available(),
    verbose=True
)

with Timer() as t:
    bivae.fit(train_set)
print("Took {} seconds for training.".format(t))
```

```python
with Timer() as t:
    all_predictions = predict_ranking(bivae, train, usercol='userID', itemcol='itemID', remove_seen=True)
print("Took {} seconds for prediction.".format(t))
```
